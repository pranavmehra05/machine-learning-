# Split Dataset & Feature Scaling

```python
#splitting dataset into independent and dependent datasets
x = df.iloc[:, 2:].values
y = df.iloc[:, 1].values
```

```python
#splitting datsets into training(75%) and testing(25%)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#print data
x_train
```

# Machine Learning Model Building


KNN

```python
#build a KNN model
from sklearn.neighbors import KNeighborsClassifier
knclassifier = KNeighborsClassifier()
knclassifier.fit(x_train, y_train)

#make use of trained model to make predictions
knn_prediction = knclassifier.predict(x_test)
```

Support Vector Classifier

```python
#build a svm model
from sklearn.svm import SVC
svclassifier = SVC()
svclassifier.fit(x_train, y_train)

svc_prediction = svclassifier.predict(x_test)
```

Decision Tree Classifier

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier(criterion='entropy', random_state=51)
dtclassifier.fit(x_train, y_train)

dt_prediction = dtclassifier.predict(x_test)
```

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm = confusion_matrix(y_test, knn_prediction)
print(cm)
sns.heatmap(cm, annot=True)
```

```python
cm2 = confusion_matrix(y_test, svc_prediction)
print(cm2)
sns.heatmap(cm2, annot=True)
```

```python
cm3 = confusion_matrix(y_test, dt_prediction)
print(cm3)
sns.heatmap(cm3, annot=True)
```

```python
#get accuracy score for model
from sklearn.metrics import accuracy_score
score1 = accuracy_score(y_test, knn_prediction)*100
score2 = accuracy_score(y_test, svc_prediction)*100
score3 = accuracy_score(y_test, dt_prediction)*100
print('knn model accuracy', score1,'%')
print('svc model accuracy', score2,'%')
print('dt model accuracy', score3,'%')
```

```python
import matplotlib.pyplot as plt
Accuracy = [score1, score2, score3]
plt.plot(Accuracy)
```
